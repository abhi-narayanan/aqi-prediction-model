```python
# Web srap the data and store in locally

def retrieve_html(start_year, end_year):
    
    for year in range(start_year, end_year+1):
        
        for month in range(1, 13):
            
            if month < 10:
                url = f'https://en.tutiempo.net/climate/0{month}-{year}/ws-421810.html'
                
            else:
                url = f'https://en.tutiempo.net/climate/{month}-{year}/ws-421810.html'
                
            web_data = requests.get(url).text
            
            if not os.path.exists(f'html_data/{year}'):
                os.makedirs(f'html_data/{year}')
            
            with open(f'html_data/{year}/{month}.html', 'w') as f:
                f.write(web_data)
                
```

```python
def met_data(month, year):
    # Read the files
    file_html = open(f'html_data/{year}/{month}.html', 'r')
    plain_text = file_html.read()
    
    temp_data = []
    final_data = []
    # Convert html to text
    soup = BeautifulSoup(plain_text, 'lxml')
    
    for table in soup.find_all('table', {'class' : 'medias mensuales numspan'}):
        
        for tbody in table:
            
            for tr in tbody:
                a = tr.get_text()
                temp_data.append(a)
    
    rows = len(temp_data)/15
    
    for _ in range(round(rows)):
        new_temp_data = []
        for _ in range(15):
            new_temp_data.append(temp_data[0])
            temp_data.pop(0)
            
        final_data.append(new_temp_data)
        
    # Remove the header row and the last row
    final_data.pop(len(final_data)-1)
    final_data.pop(0)
    
    return final_data
```

```python
def avg_data():
    average = []
    year_list = ['2013', '2014', '2015', '2016', '2017', '2018']
    
    for year in year_list:
        
        for rows in pd.read_csv(f'AQI/aqi{year}.csv', chunksize= 24):
            data = []
            add_var = 0
            
            df = pd.DataFrame(data = rows)
            
            for _, row in df.iterrows():
                data.append(row['PM2.5'])
                
            for i in data:
                if type(i) is float or type(i) is int:
                    add_var += i
                    
                elif type(i) is str:
                    if i != 'NoData' and i != 'PwrFail' and i != 'InVld' and i != '---':
                        add_var += float(i)
                
            avg = add_var/24
            
            average.append(avg)
    
    return average
```

```python
import csv
```

```python
if not os.path.exists('real_data'):
    os.makedirs('real_data')
    
for year in range(2013, 2019):
    final_data = []
    
    with open(f'real_data/real_{year}.csv', 'w') as f:
        data = csv.writer(f, dialect = 'excel')
        data.writerow(['Day','T', 'TM', 'Tm', 'SLP', 'H', 'PP','VV', 'V', 'VM', 'VG', 'RA', 'SN', 'TS', 'FG','PM 2.5'])
        
    for month in range(1, 13):
        temp = met_data(month, year)
        final_data += temp
    
    pm = avg_data()
    
    for i in range(len(final_data)-1):
        final_data[i].insert(15, pm[i])
        
    with open(f'real_data/real_{year}.csv', 'a') as f:
        data = csv.writer(f, dialect = 'excel')
        
        for row in final_data:
            data.writerow(row)
```

```python
def year_range(start, end):
    
    df_list = []
    
    for year in range(start, end):
        df = pd.read_csv(f'real_data/real_{year}.csv')
        df_list.append(df)
    
    return pd.concat(df_list, ignore_index=True)
```

```python
# Feature Scaling

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df.drop('PM 2.5', axis = 1))
```

```python
# Train test split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_data, df['PM 2.5'], test_size=0.2, random_state=42)
```

```python
# Import Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_percentage_error(y_test, predictions)
```

```python
r2 = r2_score(y_test, predictions)
r2
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
# Cross validation
from sklearn.model_selection import cross_val_score
score = cross_val_score(lr,scaled_data, df['PM 2.5'],cv=5)
```

```python
score
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor
```

```python
predictions_rfc = rfc.predict(X_test)
```
